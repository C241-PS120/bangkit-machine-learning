```python
import os
from PIL import Image
from rembg import remove

# Path ke direktori asal dan tujuan
source_directory = '/content/drive/MyDrive/Bangkit 2024 Batch 1/Healthy'
destination_directory = '/content/drive/MyDrive/Bangkit 2024 Batch 1 remove BG/Healthy'

# Buat direktori tujuan jika belum ada
if not os.path.exists(destination_directory):
    os.makedirs(destination_directory)

# Loop melalui semua file dalam direktori asal
for filename in os.listdir(source_directory):
    # Jika file memiliki format .jpg atau .png
    if filename.endswith('.jpg') or filename.endswith('.png'):
        # Path lengkap file asal dan tujuan
        source_path = os.path.join(source_directory, filename)
        destination_path = os.path.join(destination_directory, filename)

        # Baca gambar
        with Image.open(source_path) as img:
            # Hapus latar belakang
            img_no_bg = remove(img)

            # Konversi gambar ke mode RGB jika dalam mode RGBA
            if img_no_bg.mode == 'RGBA':
                img_no_bg = img_no_bg.convert('RGB')

            # Simpan gambar tanpa latar belakang di direktori tujuan
            img_no_bg.save(destination_path, format='JPEG')
            print(f"File {filename} telah dihapus latar belakangnya dan disimpan di {destination_directory}")

print("Proses penghapusan latar belakang selesai.")
```
